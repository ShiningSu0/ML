```python
price=pd.DataFrame({'price':samsung_price})
```

```python
#bollinger band
ma_len=20
percent_bandwidth=2
price['Moving Average']=price['price'].rolling(window=ma_len).mean()
price['StdDev']=price['price'].rolling(window=ma_len).std()
price['Upper BB']=price['Moving Average']+(price['StdDev']*percent_bandwidth)
price['Lowper BB']=price['Moving Average']-(price['StdDev']*percent_bandwidth)
```

```python
plt.figure(figsize=(100,20))
plt.plot(price)
```

```python
def bb(x, w=20, k=2):
    """
    Calculate Bollinger Bands
    ubb = MA_w(x) + k * sd(x)
    mbb = MA_w(x)
    lbb = MA_w(x) - k * sd(x)
    :param x:
    :return: (ubb, mbb, lbb)
    """
    x = pd.Series(x)
    mbb = x.rolling(w).mean()
    ubb = mbb + k * x.rolling(w).std()
    lbb = mbb - k * x.rolling(w).std()
    ubb.plot(x='Date', y='UBB')
    mbb.plot(x='Date', y='MBB')
    lbb.plot(x='Date', y='LBB')
    plt.show()
```
